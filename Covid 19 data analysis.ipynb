# EDA: Covid19 Data Analysis

```python
#importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the data set
dataset= pd.read_csv("C:\\Users\\User\\PYTHON\\New folder\\covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
#Removing the useless columns
df= dataset.drop(["Lat", "Long"], axis= 1, inplace= True)
```

```python
dataset.head()
```

```python
#Aggregate the rows by country
corona_dataset_aggregated = dataset.groupby("Country/Region").sum()
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
#Visualize data related to a country
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["France"].plot()
plt.legend()
```

```python
# Calculate the first derivative of the curve
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
#Find the maximum rate of infection
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
#maximum rate for all countries
countries =list(corona_dataset_aggregated.index)
max_infection_rates = []

for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["Max infection rates"] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

```python
#creating a new dataframe
corona_data= pd.DataFrame(corona_dataset_aggregated["Max infection rates"])
corona_data
```

```python
#Import second dataset
happiness_report = pd.read_csv("C:\\Users\\User\\PYTHON\\New folder\\worldwide_happiness_report.csv")
happiness_report.head()
```

```python
#removing unnecessary columns
useless_cols= ["Overall rank","Score","Generosity","Perceptions of corruption"]
happiness_report.drop(useless_cols, axis= 1, inplace= True)
```

```python
happiness_report.head()
```

```python
#set countries as index
happiness_report.set_index("Country or region", inplace= True)
happiness_report.head()
```

```python
#check the shape of two dataframe
print(corona_data.shape)
print(happiness_report.shape)
```

```python
#join the two datasets
data = corona_data.join(happiness_report, how = "inner")
data
```

```python
#correlation between features
data.corr()
```
